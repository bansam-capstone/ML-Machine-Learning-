```python
import pandas as pd
import json
import numpy as np
```

```python
with open('data.json', "r") as file:
    data = json.load(file)
```

```python
data = [i for data in data['data'][0]['cuaca'] for i in data]
```

```python
df = pd.DataFrame(data)
```

```python
df.rename(columns={'t': 'temperature',
                   'tcc': 'cloudiness',
                    'tp': 'rain',
                    'wd_deg': 'wind_direction',
                    'wd': 'wind_direction_from',
                    'wd_to': 'wind_direction_to',
                    'ws': 'wind_speed',
                    'hu': 'humidity',
                    'vs': 'visibility',
                    'vs_text': 'visibility_text',}, inplace=True)
```

```python
df.drop(['weather','weather_desc','wind_direction_from','wind_direction_to','visibility','visibility_text','time_index','local_datetime','datetime','analysis_date','utc_datetime','image','weather_desc_en'],axis=1,inplace=True)
```

```python
columns_order = ['temperature', 'humidity', 'wind_speed', 'wind_direction', 'rain', 'cloudiness']
df = df[columns_order]
```

```python
df.to_csv('csv/json_data_converted.csv',index=False)
```
